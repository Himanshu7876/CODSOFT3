importing libraries for algorithm ,oversampling and to check score

```python
from sklearn.linear_model import LogisticRegression as lr
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score,recall_score,f1_score,confusion_matrix
from imblearn.over_sampling import SMOTE
```

dision of features and target 

```python
target = data["Class"]
features = data.drop("Class",axis=1)
```

```python
features
```

using  Synthetic Minority Over-sampling & resampling data

```python
smote=SMOTE(random_state=42)
x,y =smote.fit_resample (features,target )
```

splitting data for testing and training

```python
X_train ,X_test ,Y_train ,Y_test =train_test_split (x,y,test_size=0.2,random_state=42)
```

training model

```python
model=lr()
model.fit(X_train,Y_train)
y_pred= model.predict(X_test)
```

finding precision_score,recall_score,f1_score &confusion_matrix

```python

print("precision score is ",precision_score(Y_test,y_pred))
print("recall score is ",recall_score(Y_test,y_pred))
print("f1-score is ",f1_score(Y_test,y_pred))
print("confusion matrix :\n",confusion_matrix(Y_test,y_pred))
cf=confusion_matrix(Y_test,y_pred)
```

```python
p=precision_score(Y_test,y_pred)
r=recall_score(Y_test,y_pred)
f=f1_score(Y_test,y_pred)
```

```python
plt.figure(figsize=(8,6))
metrics=['precision','recall',"f1-score"]
scores = [p,r,f]
colors = ['skyblue', 'lightgreen', 'coral']
bars = plt.bar(metrics, scores, color=colors)
plt.title('Model Performance Metrics', fontsize=16)
plt.xlabel('Metrics', fontsize=14)
plt.ylabel('Score', fontsize=14)
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sb.heatmap(cf, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.xticks([0.5, 1.5], ['Genuine', 'Fraudulent'])
plt.yticks([0.5, 1.5], ['Genuine', 'Fraudulent'])
plt.show()
```
